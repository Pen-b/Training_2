```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generate some data

```python
xy1 = np.random.normal(0,1.5,size=(200,2))
xy2 = np.random.normal(3,1,size=(100,2))
xy3 = np.random.normal(0,1,size=(100,2))

xy3[:,0] = xy3[:,0]-5.0

plt.scatter(xy1[:,0], xy1[:,1], color='red')
plt.scatter(xy2[:,0], xy2[:,1], color='blue')
plt.scatter(xy3[:,0], xy3[:,1], color='cyan')
```

```python
blues = np.concatenate((xy2, xy3), axis=0)
reds = xy1
X = np.concatenate((blues, reds), axis=0)
```

```python
y = list(np.ones(200))+list(np.zeros(200))
```

```python
from sklearn import linear_model as lm
import sklearn
import sklearn.metrics as metrics
```

## Log Regression fit

```python
# create CV LR model
clf = lm.LogisticRegressionCV()
# fit to our data
clf.fit(X,y)
```

```python
def plot_decision_boundary(pred_func):
    # Set min and max values and give it some padding
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = 0.01
    
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    # Predict the function value for the whole gid
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
```

```python
#fit model and print results / plt
pred = clf.predict(X)


plot_decision_boundary(lambda x: clf.predict(x))
print('Accuracy:%.2f' %metrics.accuracy_score(y,pred))
print(metrics.confusion_matrix(y, pred))
```

## Neural Network

```python
# import the Multi_Layer_Perceptron 
from sklearn.neural_network import MLPClassifier
```

```python
# model params, 3 layers with 30 neurons
mlp = MLPClassifier(hidden_layer_sizes=(150,150))
```

```python
mlp.fit(X,y)
```

```python
# make predictions
predictions = mlp.predict(X)
```

```python
plot_decision_boundary(lambda x: mlp.predict(x))

print('Accuracy:%.2f' %metrics.accuracy_score(y,predictions))
print(metrics.confusion_matrix(y, predictions))
```
